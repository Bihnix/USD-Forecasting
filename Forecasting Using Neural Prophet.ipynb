```python
import pandas as pd 
import requests
from neuralprophet import NeuralProphet, set_log_level
# Disable logging messages unless there is an error
set_log_level("ERROR")

import warnings
warnings.filterwarnings('ignore')

url = 'https://dashboard-api.tgju.org/v1/tv/history?symbol=PRICE_DOLLAR_RL&resolution=1D&from=1681492565&to=1712678305'

res = requests.get(url=url)
res_j = res.json()
res_j.pop('s')
res_j.pop('v')
df = pd.DataFrame.from_dict(res_j)
df['t'] = pd.to_datetime(df['t'],unit='s')
df.to_csv('df.csv')
dfs = df[['t','c']].copy()
dfs.rename(columns={'t':'ds','c':'y'}, inplace= True)
dfs = dfs[dfs['ds'] >= '2017-01-01'].reset_index().drop(columns='index')
dfs = dfs.drop_duplicates(subset=['ds'])
dfs
```

```python
plt = dfs.plot(x="ds", y="y", figsize=(15, 5))
```

```python
m = NeuralProphet(seasonality_mode="additive",
                  learning_rate=0.1,
                  n_lags=10,  
                  n_forecasts=30,
                  quantiles=[0.05, 0.95],
                  daily_seasonality=False,
                  epochs=100,)
m.set_plotting_backend("plotly-static")  # show plots correctly in jupyter notebooks
m = m.add_country_holidays("IR")
df_train, df_test = m.split_df(df=dfs, freq="D", valid_p=0.2)

# metrics_train = m.fit(df_train, freq="D")
# metrics_test = m.test(df=df_test)
metrics = m.fit(df=df_train, freq="D", validation_df=df_test, progress="plot")

df_future = m.make_future_dataframe(dfs, n_historic_predictions=True,periods=365)
forecast = m.predict(df_future)
```
